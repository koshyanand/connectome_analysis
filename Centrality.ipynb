```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import igraph as ig
import networkx as nx
from utils import get_max_flow,load_nx_graph, load_graph, create_array
from methods import get_centrality, get_tuple, get_label, map_edges
import math
from networkx.algorithms import closeness_centrality
from networkx.algorithms.link_analysis.pagerank_alg import pagerank
```

```python
print("pagerank")
c_values, arg_sorted_values, mean = get_centrality(chimp_graph, "pagerank", 5)
output = get_tuple(c_values, arg_sorted_values, True)
print(mean)
print(output)

h_values, arg_sorted_values, mean = get_centrality(human_graph, "pagerank", 5)
output = get_tuple(h_values, arg_sorted_values, False)
print(mean)
print(output)


plt.plot(list(range(1,73)), h_values, label = "Human")
plt.plot(list(range(1,73)), c_values, label = "Chimpanzee")
plt.xlabel('Node')
plt.ylabel('pagerank')

plt.legend()
plt.savefig('pagerank.png')
plt.show()
```
